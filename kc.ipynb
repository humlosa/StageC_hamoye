```python
laptops_df['model'], laptops_df['spec']=laptops_df['Title'].str.split('-', 1).str
del laptops_df['Title']
laptops_df.head()
```

```python
laptops_df['RAM'], laptops_df['HDD']=laptops_df['spec'].str.split('(', 1).str
del laptops_df['spec']
laptops_df.head()
```

```python
laptops_df['RAM'], laptops_df['Hard disk']=laptops_df['HDD'].str.split('/', 1).str
del laptops_df['HDD']
laptops_df.head()
```

```python
laptops_df.rename(columns={'price': 'Model_Price', 'Rating': 'Product_rating', 'model':'Product_Name'},inplace=True)
laptops_df.head()
```
